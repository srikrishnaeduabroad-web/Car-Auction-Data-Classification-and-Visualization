Data about Cars from an auction website

Here‚Äôs a Python class Car with the specified attributes and a unique identifier set using id(self):

```python
class Car:
    def __init__(self, date, model, company, car_type, rating, price, mileage):
        self.date = date                    # Manufactured date
        self.model = model                  # Car‚Äôs model name
        self.company = company              # Manufacturer name
        self.type = car_type                # Car‚Äôs type
        self.rating = rating                # Rating on the auction website
        self.price = price                  # Car‚Äôs price in USD
        self.mileage = mileage              # Car‚Äôs mileage in miles
        self.unique_id = id(self)           # Unique identifier using memory ID

    def __str__(self):
        return f"{self.unique_id},{self.date},{self.model},{self.company},{self.type},{self.rating},{self.price},{self.mileage}"
```

```python
car1 = Car("2020-03-15", "Camry", "Toyota", "Sedan", 4.7, 18000, 25000)
print(car1)
```

Create two subclasses: ElectricCar and NonElectricCar, both inheriting from the Car clas

```python
class ElectricCar(Car):
    def __init__(self, date, model, company, rating, price, mileage):
        super().__init__(date, model, company, "Electric", rating, price, mileage)


class NonElectricCar(Car):
    def __init__(self, date, model, company, rating, price, mileage):
        super().__init__(date, model, company, "Non-Electric", rating, price, mileage)
```

The constructors in the ElectricCar and NonElectricCar classes should:
a. Take input in this order: date,model,company,rating,price,mileage
b. Use super() to call the constructor of the parent Car class, passing the type attribute as
"Electric" for ElectricCar and "non-Electric" for NonElectricCar with rest of the
input


```python
e_car = ElectricCar("2022-01-01", "Model S", "Tesla", 4.9, 85000, 10000)
ne_car = NonElectricCar("2020-06-15", "Corolla", "Toyota", 4.6, 18000, 30000)

print(e_car)
print(ne_car)
```

Create three additional subclasses: Telsa, Ford, BMW, inheriting from ElectricCar and
NonElectricCar, NonElectricCar respectively.

```python
class Tesla(ElectricCar):
    def __init__(self, date, model, rating, price, mileage):
        super().__init__(date, model, "Tesla", rating, price, mileage)


class Ford(NonElectricCar):
    def __init__(self, date, model, rating, price, mileage):
        super().__init__(date, model, "Ford", rating, price, mileage)


class BMW(NonElectricCar):
    def __init__(self, date, model, rating, price, mileage):
        super().__init__(date, model, "BMW", rating, price, mileage)
```

The constructors in the Telsa, Ford, and BMW classes should:
a. Take input in this order: date,model,rating,price,mileage
b. Set the company as class name (i.e. company = "Ford" if class is Ford).
c. Use super() to call the constructor of the parent class with the rest of the input.

```python
df = pd.read_csv('data.csv')
```

Convert appropriate columns to numeric

```python
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['mileage'] = pd.to_numeric(df['mileage'], errors='coerce')
```

Python Code for Data Cleaning and Exploration (CRISP-DM Step)


1.    Basic Exploration



```python
print("üìä Dataset shape:", df.shape)
display(df.head())          # Show first 5 rows
display(df.tail())          # Show last 5 rows
df.info()                   # Data types and non-null count
```

2.   Check for Missing Values



```python
print("\nMissing values (using .isna()):")
print(df.isna().sum())  # Check missing values per column using isna()
print("\nMissing values (using .isnull()):")
print(df.isnull().sum())  # .isnull() is equivalent to .isna(), used here for confirmation
```

3. Explore Mode and Mean



```python
# These values are used to fill missing values depending on the column type
print("\nMode of 'company':", df['company'].mode()[0])  # Most common company name
print("Mean of 'price':", df['price'].mean())           # Average price for filling missing price
```

4. Fill Missing Values Based on Data Type

```python
# Loop through each column to fill missing values:
# - If it's a string/object (categorical), use the most frequent value (mode)
# - If it's numeric, use the average value (mean)
for col in df.columns:
    if df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0], inplace=True)  # Fill with mode for categorical
    else:
        df[col].fillna(df[col].mean(), inplace=True)     # Fill with mean for numeric
```

5. Final Check ‚Äì Ensure No Missing Values Remain

```python
print("\n‚úÖ Missing values after filling:")
print(df.isnull().sum())  # Should print 0 for all columns if everything was filled
```

Visualization Code (All 7 Types)

```python
# Convert 'date' column to datetime if not already
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
# 1Ô∏è‚É£ Bar Chart ‚Äì Average Price by Car Company
# Which company has higher average prices
plt.figure(figsize=(10, 6))
df.groupby('company')['price'].mean().sort_values(ascending=False).plot(kind='bar')
plt.title('Average Price by Company')
plt.xlabel('Company')
plt.ylabel('Average Price (USD)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 2Ô∏è‚É£ Boxplot ‚Äì Price Distribution by Car Type
# How price varies within car types (outliers visible)
plt.figure(figsize=(8, 6))
sns.boxplot(x='type', y='price', data=df)
plt.title('Price Distribution by Car Type')
plt.xlabel('Type')
plt.ylabel('Price')
plt.show()
```

```python
# 3Ô∏è‚É£ Line Plot ‚Äì Price & Mileage Over Time
# Price and mileage trends over time
df_sorted = df.sort_values(by='date')
plt.figure(figsize=(12, 6))
plt.plot(df_sorted['date'], df_sorted['price'], label='Price', color='blue')
plt.plot(df_sorted['date'], df_sorted['mileage'], label='Mileage', color='green')
plt.title('Car Price and Mileage Over Time')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Ensure 'date' is datetime type
df['date'] = pd.to_datetime(df['date'])

# Create a 'month' column in Year-Month format
df['month'] = df['date'].dt.to_period('M').dt.to_timestamp()

# Group by 'month' and calculate mean for price and mileage
monthly_avg = df.groupby('month')[['price', 'mileage']].mean().reset_index()

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(monthly_avg['month'], monthly_avg['price'], label='Average Price', color='blue')
plt.plot(monthly_avg['month'], monthly_avg['mileage'], label='Average Mileage', color='green')
plt.title('Average Car Price and Mileage by Month')
plt.xlabel('Month')
plt.ylabel('Average Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 4Ô∏è‚É£ Scatter Plot ‚Äì Price vs. Mileage colored by Company
# Price vs mileage pattern, colored by brand
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='mileage', y='price', hue='company', alpha=0.7)
plt.title('Price vs Mileage by Company')
plt.xlabel('Mileage (miles)')
plt.ylabel('Price (USD)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# 5Ô∏è‚É£ Histogram ‚Äì Distribution of Ratings
# Distribution (frequency) of car ratings
plt.figure(figsize=(8, 6))
df['rating'].hist(bins=10, edgecolor='black')
plt.title('Distribution of Car Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 6Ô∏è‚É£ Pie Chart ‚Äì Car Type Distribution (Electric vs Non-Electric)
# Proportion of electric vs non-electric cars
plt.figure(figsize=(6, 6))
df['type'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set2"))
plt.title('Car Type Distribution')
plt.ylabel('')
plt.show()
```

```python
# 7Ô∏è‚É£ Heatmap ‚Äì Correlation between Price, Rating, and Mileage
# Numeric feature relationships (correlations)
plt.figure(figsize=(8, 6))
correlation = df[['price', 'rating', 'mileage']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap (Price, Rating, Mileage)')
plt.show()
```
